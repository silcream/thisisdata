## 투수의 연봉 예측

```python
pd.options.mode.chained_assignment = None
# pandas 형태로정의된 데이터출력할 때 float모양으로 출력되게 함
```

```python
def standard_scaling(df, scale_columns):
    for col in scale_columns:
        series_mean = df[col].mean()
        series_std = df[col].std()
        df[col] = df[col].apply(lambda x: (x-series_mean)/series_std) #z-score 구하는 공식
    return df
```

```python
# 피처 각각에 대한 스케일링 
scale_columns = ['승', '패', '세', '홀드', '블론', '경기', '선발', '이닝', '삼진/9',
       '볼넷/9', '홈런/9', 'BABIP', 'LOB%', 'ERA', 'RA9-WAR', 'FIP', 'kFIP', 'WAR', '연봉(2017)']
picher_df = standard_scaling(picher, scale_columns)
```

```python
picher_df = picher_df.rename(columns={'연봉(2018)': 'y'})
picher_df.head(5) #데이터들의 범위가 비슷해짐!! 
```

## 어떤 피처가 가장 영향력이 강할까???

#### 계수를 보고 영향력이 가장 큰 것

#### 회귀분석의 통계적 분석
#### 사이킷런과  stats 라이브러리

#### 통계기반의 회귀분석 - 각각의 변수에 대한 p-value 확인
#### statsmodels에 OLS함수 넣기
#### r-square : 직선이 데이터를 얼마나 잘 설명하고 있는가
#### adj : 수식적으로 보정함(비슷)

#### f 통계량 : 모델 자체가 얼마나 잘 수행되었는지 검증

#### p-value가 0에 가까울수록 유효한 검증
#### 연봉 한화 등이 0에 가까움

```python
import statsmodels.api as sm

X_train = sm.add_constant(X_train)
model = sm.OLS(y_train, X_train).fit()
model.summary()
```

## 예측모델 평가

```python
X = picher_df[picher_df.columns.difference(['선수명', 'y'])]
y = picher_df['y']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=19)
```

```python
lr = linear_model.LinearRegression()
model = lr.fit(X_train, y_train)
```

### R2 score

```python
print(model.score(X_train, y_train)) 
print(model.score(X_test, y_test)) 
```

### RMSE score

```python
# 0~1이 아닌 
# 거리 합한거에 루트 씌워준 거에 대한 점수 계산
# 보통 r2 score와 rmse score 동시에 보고 판단함
```

```python
y_predictions = lr.predict(X_train)
print(sqrt(mean_squared_error(y_train, y_predictions)))
y_predictions = lr.predict(X_test)
print(sqrt(mean_squared_error(y_test, y_predictions)))
```

## 예측성능 높이기 : 다중공선성 확인

```python
# 대각선이 아닌데도 1에 가까움(상관도 높음)
#문제 :
# y = 내가 졸린 정도
# x1 : 커피 얼마나 마셨는지 (카페인)
# x2 : 콜라 얼마나 마셨는지 (카페인)
# x3 : 레드불 얼마나 마셨는지 (카페인)
#
# -> 서로간에 영향을 미치는 피처이다 -> 예측 성능에 영향을 줌 : 다중공선성이 발생했다
# 
# 다중공선성 : 변수간의 상관관계가 높아서 분석에 부정적인 영향을 미치는 것
# 피처간의 연관성을 고려했을 때 VIF가 높으면 (15정도면 높음) 공선성에 문제가있다 판단 -> 제거하기
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

#공선성 검증하기

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
vif.round(1)
```
